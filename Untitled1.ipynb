```python
train = df[df['Age'].notna()]
train.head()
```

```python
train.shape
```

```python
test =df[df['Age'].isna()]
test.head()
```

```python
test.shape
```

```python
cols =['Survived','Pclass','SibSp','Parch','Fare']
x_train = train[cols]
x_train.head()
```

```python
y_train = train['Age']
y_train.head()
```

```python
x_test = test[cols]
x_test.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
df.loc[df['Age'].isna(),'Age']=y_pred
```

```python
train = df[df['Embarked'].notna()]
test = df[df['Embarked'].isna()]

x_train = train[cols]
y_train = train['Embarked']

x_test = test[cols]

#crear un modelo knn
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)
y_pred
```
