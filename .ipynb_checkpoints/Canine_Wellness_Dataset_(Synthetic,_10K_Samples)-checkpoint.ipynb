```python
import warnings
warnings.simplefilter('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import itertools
from sklearn.model_selection import cross_val_score, cross_val_predict, KFold
from sklearn.preprocessing import StandardScaler, OrdinalEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc

# Configuración de estilo para gráficos
sns.set(style='whitegrid', palette='muted', font_scale=1.1)
%matplotlib inline
```

```python
df = pd.read_csv('synthetic_dog_breed_health_data.csv')
df.info()
print(df.isnull().sum())
df.head()
```

```python
df.describe()
```

```python
numeric_cols = ['Age', 'Weight (lbs)', 'Daily Walk Distance (miles)', 'Hours of Sleep',
                'Play Time (hrs)', 'Annual Vet Visits', 'Average Temperature (F)']
categorical_cols = ['Breed', 'Breed Size', 'Sex', 'Daily Activity Level', 'Owner Activity Level',
                    'Food Brand', 'Spay/Neuter Status', 'Diet', 'Other Pets in Household',
                    'Medications', 'Seizures', 'Healthy']
for col in categorical_cols:
    print(f"\n{df[col].value_counts()}")
```

```python
# Histogramas para numéricas
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[col], kde=True)
    plt.title(f'Histograma de {col}')
plt.tight_layout()
plt.show()

# Boxplots para numéricas
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot de {col}')
plt.tight_layout()
plt.show()

# Matriz de correlación
plt.figure(figsize=(10, 8))
correlation_matrix = df[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlación de Variables Numéricas')
plt.show()

combinations = list(itertools.combinations(numeric_cols, 2))
n_plots = len(combinations)  # 21 combinaciones
rows = 5
cols = 5
plt.figure(figsize=(20, 20))
for i, (x_col, y_col) in enumerate(combinations, 1):
    plt.subplot(rows, cols, i)
    sns.scatterplot(data=df, x=x_col, y=y_col, hue='Healthy', style='Healthy')
    plt.title(f'{x_col} vs {y_col}')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
plt.tight_layout()
plt.show()


# Gráficas de conteo para categóricas seleccionadas
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Daily Activity Level', hue='Healthy')
plt.title('Conteo de Daily Activity Level por Healthy')
plt.xticks(rotation=45)
plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Breed Size', hue='Healthy')
plt.title('Conteo de Breed Size por Healthy')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Verificar y eliminar filas con y nula ---
print(f"Nulos en Healthy: {df['Healthy'].isnull().sum()}")
df = df.dropna(subset=['Healthy'])
print(f"Nulos en Healthy después de eliminación: {df['Healthy'].isnull().sum()}")
```

```python
# --- Seleccionar columnas relevantes ---
features = [
    'Breed', 'Breed Size', 'Sex', 'Age', 'Weight (lbs)', 'Spay/Neuter Status',
    'Daily Activity Level', 'Diet', 'Food Brand', 'Daily Walk Distance (miles)',
    'Other Pets in Household', 'Medications', 'Seizures', 'Hours of Sleep',
    'Play Time (hrs)', 'Owner Activity Level', 'Annual Vet Visits',
    'Average Temperature (F)'
]
X = df[features]
y = df['Healthy']
```

```python
# --- Definir órdenes ordinales ---
breed_order = [
    'Australian Shepherd', 'Dachshund', 'Chihuahua', 'Siberian Husky',
    'Boxer', 'Labrador Retriever', 'Bulldog', 'Rottweiler',
    'German Shepherd', 'Golden Retriever', 'Poodle', 'Doberman',
    'Great Dane', 'Beagle', 'Yorkshire Terrier'
]
breed_size_order = ['Small', 'Medium', 'Large']
daily_activity_order = ['None', 'Low', 'Moderate', 'Active', 'Very Active']
owner_activity_order = ['None', 'Low', 'Moderate', 'Active', 'Very Active']
food_brand_order = [
    'Wellness', 'Purina', 'Iams', 'Blue Buffalo',
    'Royal Canin', 'Nutro', 'Pedigree', "Hill's Science", 'Special'
]
```

```python
# --- Tratamiento de Nulos ---
# Numéricas continuas (promedio)
numeric_mean_cols = ['Weight (lbs)', 'Daily Walk Distance (miles)', 'Play Time (hrs)', 'Average Temperature (F)']
mean_imputer = SimpleImputer(strategy='mean')
X[numeric_mean_cols] = mean_imputer.fit_transform(X[numeric_mean_cols])

# Numéricas discretas (mediana)
numeric_median_cols = ['Age', 'Hours of Sleep', 'Annual Vet Visits']
median_imputer = SimpleImputer(strategy='median')
X[numeric_median_cols] = median_imputer.fit_transform(X[numeric_median_cols])

# Categóricas nominales (moda)
nominal_mode_cols = ['Sex', 'Other Pets in Household', 'Medications', 'Seizures']
mode_imputer = SimpleImputer(strategy='most_frequent')
X[nominal_mode_cols] = mode_imputer.fit_transform(X[nominal_mode_cols])

# Categóricas nominales (nueva clase para Spay/Neuter Status y Diet)
constant_imputer = SimpleImputer(strategy='constant', fill_value='None')
X['Spay/Neuter Status'] = constant_imputer.fit_transform(X[['Spay/Neuter Status']]).ravel()

diet_imputer = SimpleImputer(strategy='constant', fill_value='Unknown')
X['Diet'] = diet_imputer.fit_transform(X[['Diet']]).ravel()

# Categóricas ordinales (moda para Breed, Breed Size, Food Brand)
ordinal_mode_cols = ['Breed', 'Breed Size', 'Food Brand']
ordinal_imputer = SimpleImputer(strategy='most_frequent')
X[ordinal_mode_cols] = ordinal_imputer.fit_transform(X[ordinal_mode_cols])

# Categóricas ordinales (nueva clase para Daily Activity Level y Owner Activity Level)
activity_imputer = SimpleImputer(strategy='constant', fill_value='None')
X['Daily Activity Level'] = activity_imputer.fit_transform(X[['Daily Activity Level']]).ravel()
X['Owner Activity Level'] = activity_imputer.fit_transform(X[['Owner Activity Level']]).ravel()

# Verificar que no hay más nulos
X.isnull().sum()
```

```python
# --- Convertir a Numéricos ---
# Codificación ordinal
ordinal_cols = ['Breed', 'Breed Size', 'Daily Activity Level', 'Owner Activity Level', 'Food Brand']
ordinal_mappings = [
    breed_order, breed_size_order, daily_activity_order, owner_activity_order, food_brand_order
]
ordinal_encoder = OrdinalEncoder(categories=ordinal_mappings)
X[ordinal_cols] = ordinal_encoder.fit_transform(X[ordinal_cols])

# One-hot encoding para nominales
nominal_cols = ['Sex', 'Other Pets in Household', 'Medications', 'Seizures', 'Spay/Neuter Status', 'Diet']
X = pd.get_dummies(X, columns=nominal_cols, drop_first=True, dtype=int)

# --- Escalado de Datos ---
scaler = StandardScaler()

X[:] = scaler.fit_transform(X[:])

# Mostrar información y primeras filas
print(X.info())
X.head()
```

```python
# Definir modelos
rf_model = RandomForestClassifier(random_state=60)
svm_model = SVC(probability=True, random_state=30)

# Configurar k-fold
kf = KFold(n_splits=5, shuffle=True, random_state=60)

def evaluate_model(model, X, y, model_name):
    start_time = time.time()

    # Accuracy con cross-validation
    accuracy_scores = cross_val_score(model, X, y, cv=kf, scoring='accuracy')

    # Predicciones para otras métricas
    y_pred = cross_val_predict(model, X, y, cv=kf)

    # Calcular métricas
    accuracy = accuracy_scores.mean()
    precision = precision_score(y, y_pred, pos_label='Yes')
    recall = recall_score(y, y_pred, pos_label='Yes')
    f1 = f1_score(y, y_pred, pos_label='Yes')

    # Tiempo de entrenamiento
    end_time = time.time()
    training_time = end_time - start_time

    print(f"\nMétricas para {model_name}:")
    print(f"Accuracy (k-fold): {accuracy:.4f} ± {accuracy_scores.std():.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"Tiempo de entrenamiento: {training_time:.2f} segundos")

    # Matriz de confusión
    cm = confusion_matrix(y, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Matriz de Confusión - {model_name}')
    plt.xlabel('Predicho')
    plt.ylabel('Real')
    plt.show()

    return accuracy, precision, recall, f1, y_pred

# Evaluar modelos
rf_metrics = evaluate_model(rf_model, X, y, "Random Forest")
svm_metrics = evaluate_model(svm_model, X, y, "SVM")
```

```python
# Curvas ROC con cross-validation
plt.figure(figsize=(8, 6))
for model, model_name in [(rf_model, "Random Forest"), (svm_model, "SVM")]:
    tprs = []
    aucs = []
    mean_fpr = np.linspace(0, 1, 100)

    for train_idx, test_idx in kf.split(X):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        model.fit(X_train, y_train)
        y_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_prob, pos_label='Yes')
        tprs.append(np.interp(mean_fpr, fpr, tpr))
        aucs.append(auc(fpr, tpr))

    mean_tpr = np.mean(tprs, axis=0)
    mean_auc = np.mean(aucs)
    plt.plot(mean_fpr, mean_tpr, label=f'{model_name} (AUC = {mean_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC - Cross-Validation')
plt.legend(loc='lower right')
plt.show()

# Comparación de métricas
metrics_df = pd.DataFrame({
    'Métrica': ['Accuracy', 'Precision', 'Recall', 'F1-Score'],
    'Random Forest': rf_metrics[:4],
    'SVM': svm_metrics[:4]
})
plt.figure(figsize=(10, 6))
metrics_df.plot(x='Métrica', kind='bar', figsize=(10, 6))
plt.title('Comparación de Métricas: Random Forest vs SVM')
plt.ylabel('Score')
plt.xticks(rotation=0)
plt.legend()
plt.show()
```
